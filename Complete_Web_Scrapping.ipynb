<article style="background-color:skyblue;font:red;">Save Large file from internet</article>
<article style="background-color:gold;font:red;">인터넷에 있는대용량 파일을 내 PC 로 저장 하기</article>

```python
from urllib.request import urlopen
url="https://www.python.org/ftp/python/3.7.3/python-3.7.3-macosx10.9.pkg" #this is mac version
#url="https://www.python.org/ftp/python/3.7.2/python-3.7.2/python-3.7.2.exe" #Always copy sites from download site
filename=url.split("/")[-1]
BUFFSIZE=256*1024  # Download speed 256 kbps

try:
    with urlopen(url) as uopen: #open url 
        with open(filename,"wb") as uwrite: #copy file 
            buf=uo.read(BUFFSIZE) #copy speed
            while (buf):  #
                uwrite.write(buf)
                buf=uopen.read(BUFFSIZE)
except Exception as e:
    print(e)
```

### Beautifulsoup 기본 사용
다음은 Beautifulsoup를 이용하여 웹사이트로부터 HTML을 가져와 문자열로 만들어 이용하는 예제임
* h1 태그를 접근하기 위해 html-body-h1 구조를 사용하여 soup.html.body.h1 이런식으로 이용하게 됨.
* p 태그는 두개가 있어 soup.html.body.p 한 후 next_sibling을 두번 이용하여 다음 p를 추출. 한번만 하면 그 다음 공백이 추출됨.
* HTML 태그가 복잡한 경우 이런 방식으로 계속 진행하기는 적합하지 않음.

```python
#import bs4 (Beautiful Library)
from bs4 import BeautifulSoup
html="""
<html>
    <body>
        <h1>I am Krish.저는 네팔에서 왔습니다</h1>
        <p>I love Data Analysis and Computer Networking</p>
        <p>다음은 Beautifulsoup를 이용하여 웹사이트로부터 HTML을 가져와 문자열로 만들어 이용하는 예제임</p>
    </body>
</html>
"""
#Analysis Html
soup=BeautifulSoup(html,"html.parser")
h1=soup.html.body.h1
p1=soup.html.body.p
p2=p1.next_sibling.next_sibling

#Printing The text stored 
print("H1 is :>>\n",h1.string) #with the help of .string html tag is avoided

print("\nFirst Paragraph is \n",p1.string)

print("\nSecond Paragraph is \n",p2.string)

```

<article style="background-color:lightblue;font-size:18px;color:red">We Use fine_all() Method to find other tags<br><font color=blue>This Code find all link fils in Html sites</font></article>

```python
from bs4 import BeautifulSoup
html="""
<html>
    <body>
        <h1>I am Krish.저는 네팔에서 왔습니다</h1>
        <h1>I have been living in Korea Since 2011</h1>
        <p>I love Data Analysis and Computer Networking</p>
        <p>다음은 Beautifulsoup를 이용하여 웹사이트로부터 HTML을 가져와 문자열로 만들어 이용하는 예제임</p>
        <ul>
            <li><a href="http://www.google.com">Google</a></li>
            <li><a  href="http://www.facebook.com">Facebook</a></li>
        </ul>
        <ol>
            <li><a href="http://www.auction.co.kr">Auction</a></li>
            <li><a  href="http://www.gmarket.co.kr">Gmarket</a></li>
        </ol>
        
    </body>
</html>
"""
soup = BeautifulSoup(html,"html.parser")
links=soup.find_all("a") #all link files


for a in links:
        href=a.attrs['href']  # href의 속성에 있는 속성값을 추출
        text=a.string
        print(text,">",href)
```
