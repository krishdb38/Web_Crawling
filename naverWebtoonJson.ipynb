# 네이버 웹툰 리뷰 Json 파일 요청하기

```python
import json
import requests
from bs4 import BeautifulSoup as bs4
#from konlpy.tag import Twitter
#from konlpy.tag import Kkma
import csv, codecs
import pandas as pd
import os
import openpyxl
```

```python
url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?ticket=comic&templateId=webtoon&pool=cbox3&_callback=jQuery112407305799293354356_1537286021509&lang=ko&country=KR&objectId=183559_401&categoryId=&pageSize=15&indexSize=10&groupId=&listType=OBJECT&pageType=default&page=1&refresh=true&sort=new&_=1537286021511'

header = {
    #"Host": "apis.naver.com"
    "Referer": "https://comic.naver.com/comment/comment.nhn?titleId=651673&no=341",
    #"Accept-Language":"ko-KR"
}

requested = requests.get(url, headers = header) # get 방식으로 json 파일 요청하기

soup = bs4(requested.content, 'lxml', from_encoding='utf-8') # 계층적인 html 형태의 beuatifulsoup 객체로 바꿔주기

reviews = soup.select('p')[0].text

strShapedText = reviews[reviews.find('(') + 1:reviews.find(');')]

jsonShapedText = json.loads(strShapedText)
comments = jsonShapedText['result']['commentList']
for comment in comments:
        print(comment['contents'])
# pandas 객체를 사용해서 csv 파일로 만드는 부분임.
# 컴퓨터 밑에 적당한 폴더를 지정
os.chdir("c:\\r_temp\\")
df=pd.DataFrame.from_dict(comments, orient='columns')
df.to_csv('wt-review.csv',encoding='euc-kr')
```

# 여러 페이지에 걸쳐서 네이버 웹툰 리뷰 Json 파일 요청하기

```python
commentBox = []
num = 1

while 1:

    url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?ticket=comic&templateId=webtoon&pool=cbox3&_callback=jQuery112407305799293354356_1537286021509&lang=ko&country=KR&objectId=183559_401&categoryId=&pageSize=15&indexSize=10&groupId=&listType=OBJECT&pageType=default&page='+ str(num) +'&refresh=true&sort=new&_=1537286021511'

    header = {
        #"Host": "apis.naver.com"
        "Referer": "https://comic.naver.com/comment/comment.nhn?titleId=651673&no=341",
        #"Accept-Language":"ko-KR"
    }

    requested = requests.get(url, headers = header) # get 방식으로 json 파일 요청하기

    soup = bs4(requested.content, 'lxml', from_encoding='utf-8') # 계층적인 html 형태의 beuatifulsoup 객체로 바꿔주기

    reviews = soup.select('p')[0].text

    strShapedText = reviews[reviews.find('(') + 1:reviews.find(');')]

    jsonShapedText = json.loads(strShapedText)

    comments = jsonShapedText['result']['commentList']
    for comment in comments:
        commentBox.append(comment['contents'])
    
    if not len(comments):
        break
    else:
        num += 1

# csv 로 저장하기 encoding='utf-8' 넣어도 한글은 깨짐, 메모장으로 열어서 봐야함.
pd.DataFrame({"comments":commentBox}).to_csv("comments.csv")
# 엑셀로 저장하기
writer = pd.ExcelWriter('comments.xlsx', engine='xlsxwriter')
pd.DataFrame({"comments":commentBox}).to_excel(writer,'Sheet1')
writer.save()
```
