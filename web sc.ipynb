# 트립 어드바이저 크롤링

 - 각 리뷰 페이지의 url은 url1 다음에 오는 숫자에 의해 변경됨
 - while문으로 리뷰페이지들을 순차적으로 방문
 - 가져온 리뷰가 없으면(reviews_temp == []) while문 종료

```python
from bs4 import BeautifulSoup
import urllib.request as req
import datetime
import os
import pandas as pd
import openpyxl
# working directory 설정
os.chdir("c:\\r_temp\\")

# HTML 가져오기
url1="https://www.tripadvisor.co.kr/Hotel_Review-g1152744-d786773-Reviews-or"
url2="-Away_Koh_Kood_Resort-Ko_Kut_Trat_Province.html"
reviews_temp= []
reviews= []
titles = []
reviewNum = 0

while True:
    url=url1+str(reviewNum)+url2
    res = req.urlopen(url)
    soup = BeautifulSoup(res, "html.parser")
    reviews_temp = [review.text.strip() for review in soup.select("p.partial_entry")]
    
    if reviews_temp == []:
        break
    
    print(url)
    print(reviews_temp)
    
    reviews.append(reviews_temp)
    # url의 숫자는 5씩 증가한다는 것을 url에서 직접확인해야함
    reviewNum += 5
    

# xlsx 및 csv 파일로 저장
pd.DataFrame.from_dict(reviews, orient='columns').to_csv("review1.csv")
writer = pd.ExcelWriter('review1.xlsx', engine='xlsxwriter')
pd.DataFrame({"reviews":reviews}).to_excel(writer,'Sheet1')
writer.save()
```

## 자동 스크롤

 - 스크롤링 했을 때, 추가 리뷰들이 로딩되고 있음을 알려주는 css-selector 를 활용
 - try구문에서 첫 줄은 맨 바닥에 있는 css-selector, 이 css-selector를 클릭함으로써 가장 바닥 부분으로 내려가게되고 스크롤링을 한 것과 같은 효과를 낸다.
 - try구문에서 두 번째 줄은 스크롤링 했을 때 로딩되고 있음을 알려주는 css-selector, 리뷰를 모두 불러오면 이 css-selector는 나타나지 않는다.
 - 즉, 모든 리뷰를 불러왔을 때 로딩 css-selector가 나타나지 않음을 활용해서 while문을 멈춘다.
 - 스크롤링이 모두 완료되지 않았음에도 while문이 끝나는 경우가 발생, 따라서 총 10번 실패하게 되면 스크롤링을 종료하도록 해주었다.

```python
cnt = 0

while True:
    try:
        driver.find_element_by_xpath("//div[@class='loris']").click()
        driver.find_element_by_xpath("//div[@class='jfk-activityIndicator']")
        time.sleep(1)
    except:
        cnt += 1
        if cnt == 10:
            break
```

## 본격적인 리뷰 크롤링

 - 구글 리뷰페이지는 복잡한 html 구조를 가지고 있다. 실제로 확인해보면 html 태그 블록 하나에 리뷰가 10개씩 수록이 되어있음을 알 수가 있다.
 - 예를 들어, 총 300개의 리뷰가 있다면, 30개의 html tag 블록에 리뷰가 각각 10개씩 속해있다.
 - 각 블록들과 리뷰들에 대한 인덱싱은 단순히 숫자를 이용하면 되는데, 먼저 path1의 다음에 오는 숫자는 각 블록을 인덱싱한다.
 - 그리고 path2 다음에 오는 숫자는 하나의 블록에 속해있는 10개의 리뷰에 대해 인덱싱한다.
 - 모든 리뷰를 가져온 후에는 에러가 발생하므로 except 구문이 실행되고 while문이 종료된다.

```python
google_reviews = []

path1 = "//div[@id='reviewSort']//div["
path2 = "]//div[@class='gws-localreviews__general-reviews-block']//div["
path3 = "]//div[@class='jxjCjc']//div[@class='Jtu6Td']"

i = 1

while True:
    try:
        google_review_temp = []
        for j in range(1,10):
            # review 내용 가져오기
            google_review_temp.append(driver.find_element_by_xpath(path1 + str(i) + path2 + str(j) + path3).text)
        
        google_reviews.append(google_review_temp)
        i += 1
    except:
        break
        
# xlsx 및 csv 파일로 저장
pd.DataFrame.from_dict(google_reviews, orient='columns').to_csv("review2.csv")
writer = pd.ExcelWriter('review2.xlsx', engine='xlsxwriter')
pd.DataFrame({"reviews":google_reviews}).to_excel(writer,'Sheet1')
writer.save()
```
